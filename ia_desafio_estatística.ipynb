## Exercício do módulo
Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:
- Trazer a média das vendas
- Criar um gráfico de barras vertical mostrando o mês de referência e o valor
- Criar um gráfico de linhas mostrando o mês de referência e o valor.

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

### Importanto bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Criando DataFrame

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

### Solicitando informações do DataFrame

```python
df_faturamento.info()
```

### Alterando o tipo do data_ref: de object para datetime

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
df_faturamento.info()
```

### Média das vendas

```python
media_vendas = round(df_faturamento['valor'].mean(), 2)
media_vendas
```

### Ordenar o DataFrame em ordem decrescente de data

```python
df_faturamento_ordenado = df_faturamento.sort_values(by='data_ref', ascending=False)
df_faturamento_ordenado
```

### Plotando Gráfico de Barras

```python
# Tamanho do gráfico
plt.figure(figsize=(8,6))
# Gráfico
sns.barplot(x='data_ref', y='valor', data=df_faturamento_ordenado, color='red')
# Labels
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Faturamento')
# Melhorando a visualização das labels do eixo x
plt.xticks(rotation=45, ha='right')

plt.show()
```

### Plotando Gráfico de Linhas

```python
# Estilizar grid
sns.set_theme(style='darkgrid')
# Ajustar tamanho do gráfico
plt.figure(figsize=(14,10))
# Gráfico
sns.lineplot(data=df_faturamento_ordenado, x='data_ref', y='valor', color='red')
# Labels
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Faturamento')
# Melhorando a visualização das labels do eixo x
plt.xticks(df_faturamento_ordenado['data_ref'], rotation=45, ha='right')
plt.show()
```
